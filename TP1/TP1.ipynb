```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

### Ejercicio 1

```python
def met_pot(matriz, k): # Método de la potencia

  v = np.random.rand(matriz.shape[1]) # Vector aleatorio inicial
  a = []

  for i in range(k):

    Av = matriz@v
    Av_norma = np.linalg.norm(Av)
    v = Av/Av_norma

    autov_aprox = abs( np.transpose(v)@(matriz@v) / (np.transpose(v)@v) ) # Cociente de Rayleigh

    a.append(autov_aprox) # Genero vector de aproximaciones al autovalor

  return a
```

```python
A = np.random.rand(10,10) # Matriz random
k = 100 # Iteraciones
autoval_max = max(abs(np.linalg.eigvals(A))) # Autovalor maximo de A 

a = met_pot(A, k)
 
print(a)
print('\nautovalor_{0} aproximado = {1}'.format(k, a[k-1]))
print("autovalor máximo         =",autoval_max)
```

### Ejercicio 2

```python
def genMat(n,m): # Generador de matrices n x m
  A = np.random.rand(n,m)
  B = np.zeros([n,m])
  C = np.zeros([n,m])
  D = np.zeros([n,m])
  for i in range(n):
    for j in range(m):
      if i == j:  

        B[i][j] = A[i][j]

        C[i][j] = A[i][j] + 100
        
        D[i][j] = A[i][j] + 1000

      else :
        
        B[i][j] = A[i][j]
        B[j][i] = B[i][j]
        
        C[i][j] = B[i][j]
        C[j][i] = B[i][j]
        
        D[i][j] = B[i][j]
        D[j][i] = B[i][j] 

  return A, B, C, D
```

```python
A, B, C, D = genMat(100, 100)

print('A = \n{0}\nB = \n{1}\nC = \n{2}\nD = \n{3}'.format(A,B,C,D))
```

```python
# Funcion para graficar F(i) = a_i. Donde i son las iteraciones y a_i la aproximacion al autovalor 
def graf_autov_aprox(matriz,k): 

    a = met_pot(matriz,k)
    
    plt.xlim(0,k)
    plt.xlabel('iteraciones') 
    plt.ylabel('aproximacion al autovalor')
    
    for i in range(k):
        plt.plot(i, a[i], '.')
    
    plt.show()
```

### Ejercicio 3

```python
def max2_autoval(matriz): # Funcion para obtener los dos maximos autovalores
  
  w, _ = np.linalg.eigh(matriz) 

  w = sorted(abs(w))

  lambda1 = w[len(w)-1]
  lambda2 = w[len(w)-2] 
    
  return lambda1, lambda2
```

```python
def error_met_pot(matriz, k): # Funcion para obtener el vector de errores del metodo de la potencia

    lambda1, _ = max2_autoval(matriz)
    a = met_pot(matriz, k)

    error = []

    for i in range(k):
        error.append( abs(lambda1 - a[i]) ) # Genero vector de errores

    return error
```

```python
# Funcion para graficar F(i) = log(e_i). Donde e_i es el error o diferencia entre el autovalor maximo y su aproximacion
def graf_error(matriz,k):
  
  error = error_met_pot(matriz, k)
  
  plt.xlim(0, k)
  plt.ylim(-35, 10)
  plt.xlabel('iteraciones') 
  plt.ylabel('error')

  for i in range(k):
    if error[i] == 0:
      break
    else:
      plt.plot(i, mt.log(error[i]), '.', color='blue')

  
```

```python
graf_error(A,100)
```

```python
graf_error(B,100)
```

```python
graf_error(C,100)
```

```python
graf_error(D,100)
```

Aclaración: la función deja de iterar forzosamente una vez el error es 0 puesto que el logaritmo de 0 no esta definido. Luego asumimos que cuando el logaritmo del error llega a -31 y se mantiene en ese valor, este tiende a 0. Ésto es por una cuestión de la precisión con la que estamos trabajando.

```python
# Funcion para graficar F(i) = 2.log(λ1/λ2).i + log(e_0). Donde e_0 es el error inicial
def graf_error_pend(matriz, k):
    graf_error(matriz, k)
    
    error = error_met_pot(matriz, k)
    lambda1, lambda2 = max2_autoval(matriz)

    for i in range(k):
      if error[i] == 0:
        break
      else:
        plt.plot(i, (2*mt.log(lambda2/lambda1)*i) + mt.log(error[0]), '.', color='red' )
    
    
```

```python
graf_error_pend(A, 100)
```

```python
graf_error_pend(B, 100)
```

```python
graf_error_pend(C, 100)
```

```python
graf_error_pend(D, 100)
```

Observamos que la relación entre lambda 1 y lambda 2 de cada matriz esta relacionada con la velocidad de convergencia que tiene el método. Como dice el enunciado ésta velocidad está dada porque (lambda2/lambda1)^2 multiplica al error en cada iteracion, y viendo los resultados las matrices A, B convergen demasiado rápido porque el error se multiplica por aprox 0.0126 o bien se reduce al 1,26%. En cambio en la matriz C se multiplica por 0.498 y en la matriz D por 0.918 por ende éstos convergen mucho más lento. 
Es por eso que podemos ver en los gráficos con ejes fijos cómo ésta relación de los autovalores hacen que A, B converjan enseguida, C converja más lento, D más lento todavía.
Además tenemos la recta roja que esta dada por 2.log(lambda2/lambda1).i + log(error[0]), el cual representa gráficamente cómo debería evolucionar el error a partir de un error inicial. Y efectivamente se puede ver como el error que fue evolucionando (puntos azules) se aproxima a nuestra recta.

Entonces podemos concluir que mientras más cerca este lambda2 de lambda1, más cercano a 1 estará (lambda2/lambda1)^2 y más tardará en reducirse el error al multiplicarse por ésta expresión, o bien converger.
