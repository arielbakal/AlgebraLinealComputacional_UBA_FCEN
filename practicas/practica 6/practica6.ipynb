```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Ejercicio 9

```python
U = np.array([[1, -2/np.sqrt(5)],[2, 1/np.sqrt(5)]])
Z = np.array([[2,0,0],[0,1,0]])
v1, v2, v3 = np.array([2,1,2]), np.array([2,-2,-1]), np.array([1,2,-2])

# Necesito normalizar mis vectores si quiero la esfera "unitaria"
v1 = v1 / np.linalg.norm(v1)
v2 = v2 / np.linalg.norm(v2)
v3 = v3 / np.linalg.norm(v3)

V_t = np.array([v1,v2,v3])

A = U@Z@V_t
```

```python
# Esfera unitaria

# Generar puntos en la esfera unitaria
theta = np.linspace(0, 2 * np.pi, 100)
phi = np.linspace(0, np.pi, 50)

x = np.outer(np.cos(theta), np.sin(phi))
y = np.outer(np.sin(theta), np.sin(phi))
z = np.outer(np.ones_like(theta), np.cos(phi))

# Graficar los vectores y la esfera
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar los vectores
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='v1')
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='g', label='v2')
ax.quiver(0, 0, 0, v3[0], v3[1], v3[2], color='b', label='v3')

# Graficar la esfera
ax.plot_surface(x, y, z, color='gray', alpha=0.3)

# Configurar el aspecto del gráfico
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_aspect('equal')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

ax.set_title('Vectores columna de V')

# Mostrar el gráfico
plt.show()
```

```python
A = np.array([[1,-2],[2,1],[1,3]])
b = np.array([2,1,-1])
```

```python
A_pinv = np.linalg.pinv(A)
```

¿Tiene solucion?

```python
A@A_pinv@b
```

A.A†.b = b entonces sí

Hallemos la solucion minima norma-2

```python
z = A_pinv@b
print(z)
```

```python
A@z
```
