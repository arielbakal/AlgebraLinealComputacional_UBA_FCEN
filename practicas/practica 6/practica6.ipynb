#### Geometría de la descomposción SVD

```python
U = np.array([[1, -2/np.sqrt(5)],[2, 1/np.sqrt(5)]])
Z = np.array([[2,0,0],[0,1,0]])
v1, v2, v3 = np.array([2,1,2]), np.array([2,-2,-1]), np.array([1,2,-2])

# Necesito normalizar mis vectores si quiero la esfera "unitaria"
v1 = v1 / np.linalg.norm(v1)
v2 = v2 / np.linalg.norm(v2)
v3 = v3 / np.linalg.norm(v3)

V_t = np.array([v1,v2,v3])

A = U@Z@V_t
```

##### Ejercicio 12

```python
def reducir_svd(A, r):

    U, s, Vt = np.linalg.svd(A) 

    # achicar rango
    U_r = np.transpose(np.transpose(U)[:r])
    Vt_r = Vt[:r]
    s_r = s[:r]

    return np.shape(A), U_r, Vt_r, s_r
```

```python
A = np.array([[1,2,5],[2,4,7]])

reducir_svd(A, 2)
```

##### Ejercicio 13

```python
A = np.array([[1,2,5],[2,4,7],[2,3,6]])

rango, U_r, Vt_r , s_r = reducir_svd(A, 2)
```

```python
def recontruir_svd(rango, U_r, Vt_r, s_r):

    n, m = rango
    
    # agrego ceros
    for i in range(n - np.shape(U_r)[1]):
        ceros = np.zeros((1,n))
        U_r = np.transpose( np.vstack( (np.transpose(U_r) , ceros) ) )

    for j in range(m - np.shape(Vt_r)[0]):
        ceros = np.zeros((1,m))
        Vt_r = np.vstack( (Vt_r , ceros) )

    Z_r = np.zeros((n, m))

    for n in range(n):
        for m in range(m):
            if len(s_r) == 0:
                break
            if n == m:
                Z_r[n][m] = s_r[0]
                s_r = s_r[1:]

    B = U_r@Z_r@Vt_r

    return B
```

##### Ejercicio 14

```python
def comprimir_imagen_svd(imagen, r):
    n, m = np.shape(imagen)
    rango, U_r, Vt_r , s_r = reducir_svd(imagen, r)
    B = recontruir_svd(rango, U_r, Vt_r , s_r)
    plt.imshow(B, cmap="gray")
    print("Error =", np.linalg.norm(imagen - B) / np.linalg.norm(imagen) )
    print("Proporción de compresion =", (n*m) / ( (m + n)*r + r + 2 ) )
```
